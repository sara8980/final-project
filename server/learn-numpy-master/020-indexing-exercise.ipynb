```python
a.shape, a.dtype
```

```python
plt.imshow(a);
```

```python
newa = a [170:400, 250:450]
plt.imshow(newa);
newb = a [: : -1]
plt.imshow(newb);
newc = a [:,::-1]
plt.imshow(newc);

```

```python
# plt.imshow(a, cmap='Reds')
list1 = ['Reds','Greens','Blues']
for i, v in enumerate(list1):
    print(i)
    plt.title(v)
    plt.imshow(a[:,:,i], v, vmin=0, vmax=255)
    plt.colorbar()
    plt.show()
 
    
```
