## Replacement with  `slice = other_array`

```python
b = np.random.random(10)
```

```python
b
```

```python
b[3:7] = np.arange(4)
```

```python
b
```

```python
c = np.arange(15)
c
```

```python
c[5:] = c[5:] * 10
```

```python
c
```

```python
c[3:7] **= 2
```

```python
c
```
